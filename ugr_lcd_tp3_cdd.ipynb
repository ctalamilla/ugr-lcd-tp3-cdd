```python
def dataframe_summary(df: pd.DataFrame):
    summary = {}

    # Count of columns by data type
    dtype_counts = df.dtypes.value_counts().to_dict()
    summary['column_counts_by_type'] = dtype_counts

    # Analyze each column based on data type
    column_details = {}
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            # For numeric columns, calculate statistics
            stats = {
                'min': df[col].min(),
                'max': df[col].max(),
                'mean': df[col].mean(),
                'median': df[col].median(),
                'p10': df[col].quantile(0.10),
                'p90': df[col].quantile(0.90)
            }
            column_details[col] = stats
        elif pd.api.types.is_categorical_dtype(df[col]) or df[col].dtype == 'object':
            # For categorical columns, get unique values
            unique_values = df[col].unique().tolist()
            column_details[col] = {'unique_values': unique_values}

    summary['column_details'] = column_details
    return summary
```

```python
def analyze_dataframe_to_df(df):
    """
    This function takes a DataFrame as input and returns a new DataFrame with:
    - Column name
    - Data type
    - For categorical columns: unique values
    - For numerical columns: min, max, mean, median, 10th percentile, and 90th percentile
    """
    # Initialize lists to build the result DataFrame
    columns = []
    data_types = []
    unique_values = []
    min_values = []
    max_values = []
    mean_values = []
    median_values = []
    p10_values = []
    p90_values = []

    for column in df.columns:
        columns.append(column)
        data_type = df[column].dtype
        data_types.append(data_type)

        if pd.api.types.is_numeric_dtype(df[column]):
            # Numerical column statistics
            unique_values.append(None)
            min_values.append(df[column].min())
            max_values.append(df[column].max())
            mean_values.append(df[column].mean())
            median_values.append(df[column].median())
            p10_values.append(df[column].quantile(0.10))
            p90_values.append(df[column].quantile(0.90))
        elif pd.api.types.is_categorical_dtype(df[column]) or df[column].dtype == 'object':
            # Categorical column unique values
            unique_values.append(df[column].unique().tolist())
            min_values.append(None)
            max_values.append(None)
            mean_values.append(None)
            median_values.append(None)
            p10_values.append(None)
            p90_values.append(None)
        else:
            # For unsupported data types, fill with None
            unique_values.append(None)
            min_values.append(None)
            max_values.append(None)
            mean_values.append(None)
            median_values.append(None)
            p10_values.append(None)
            p90_values.append(None)

    # Create a DataFrame from the collected lists
    result_df = pd.DataFrame({
        'Column': columns,
        'Data Type': data_types,
        'Unique Values (Categorical)': unique_values,
        'Min': min_values,
        'Max': max_values,
        'Mean': mean_values,
        'Median': median_values,
        '10th Percentile': p10_values,
        '90th Percentile': p90_values
    })

    return result_df

# Example usage (assuming a DataFrame 'df' is available)
# analyzed_df = analyze_dataframe_to_df(df)
# analyzed_df
```

Emisiones (CH₄): Emisiones de metano, expresadas en masa de CH₄ (normalmente en toneladas o kilogramos). El metano es un potente gas de efecto invernadero.

Emisiones (N₂O): Emisiones de óxido nitroso, expresadas en masa de N₂O. Este gas también es un potente gas de efecto invernadero, con un impacto climático mucho mayor por tonelada en comparación con el CO₂.

Emisiones (CO₂): Emisiones de dióxido de carbono, expresadas en masa de CO₂. Es el gas de efecto invernadero más comúnmente emitido debido a actividades humanas, como la quema de combustibles fósiles.

Emisiones (CO₂eq) proveniente de CH₄ (AR5): Emisiones de metano convertidas a "equivalentes de CO₂" usando los factores de conversión del Quinto Informe de Evaluación (AR5) del IPCC (Panel Intergubernamental sobre Cambio Climático). Esto significa que las emisiones de CH₄ se han ajustado para reflejar su potencial de calentamiento global en relación con el CO₂.

Emisiones (CO₂eq) proveniente de N₂O (AR5): Emisiones de óxido nitroso, convertidas a equivalentes de CO₂ usando los factores de conversión del AR5. Esto permite comparar el impacto del N₂O en términos equivalentes al CO₂.

Emisiones (CO₂eq) (AR5): Emisiones totales en términos de "equivalentes de CO₂", calculadas sumando las emisiones de todos los gases de efecto invernadero (CH₄, N₂O, y CO₂) convertidos a CO₂eq según el AR5.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = data.loc[data['Área']!='Mundo'].reset_index(drop=True)
data
```

```python
# Para crear el gráfico , vamos a filtrar por las 10 zonas con las emisiones más elevadas de media en los años disponibles
top_10_emitters = data.groupby("Área")["Emision_MTn"].mean().nlargest(10).index
df_top_10 = data[data["Área"].isin(top_10_emitters)]
```

```python
top_10_emitters
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x="Año", y="Emision_MTn", hue="Área", data=df_top_10)

# Adding labels above each line
for area in top_10_emitters:
    area_data = df_top_10[df_top_10["Área"] == area]
    plt.text(area_data["Año"].values[-1], area_data["Emision_MTn"].values[-1], area,
             verticalalignment='bottom')

plt.title("Serie Temporal de Emisiones de CO2eq (Top 10 Emisores)")
plt.xlabel("Año")
plt.ylabel("Emisiones (MTn)")
plt.grid(True)
plt.legend(title="Área (Top 10 Emisiones)", loc="upper left")
plt.show()
```
